# Caesar III

Calculations for Caesar III (Augustus v4.0)

## Data

Determined from various sources, the augustus source itself, forum posts the augustus manual and the Caesar III manual.

```python
data = {
    "religion": {
        "lararium": {"pop": 10},
        "altar": {"pop": 50},
        "small_temple": {"pop": 750},
        "large_temple": {"pop": 3000},
        "oracle": {"pop": 500},
        "small_mausoleum": {"pop": 500},
        "large_oracle": {"pop": 750},
        "large_mausoleum": {"pop": 750},
        "nymphaeum": {"pop": 750},
    },
    "food": {
        "wheat_farm": {"pop": 160},
    },
    "housing": {
        "9x9": {"pop": 1120, "houses": }
    },
}
houses_per_workshop = 20
```

## City

Here we enter data about the current city

```python
# housing
housing_desired = {
    "9x9": 4
}
total_houses = 89
# religon (assuming one per god if building does not cover all gods)
religious_buildings = {
    "altar": 5,
    "lararium": 12,
    "small_temple": 2,
    "small_mausoleum": 1,
    "oracle": 1,
    "large_mausoleum": 1,
    "nymphaeum": 2,
}
# food
food_buildings = {
    "wheat_farm": 32
}
# production
production_buildings = {
    "pottery": 10,
    "furniture": 6
}
```

```python
desired_population = housing_desired["9x9"] * data["housing"]["9x9"]["pop"]
print(f"{desired_population=}")
```

## Calculation

Using data from various resouces, and information about the city, we calculate requirements

### Utility

Utility functions

```python
def provision(buildings, data):
    provision = 0
    for building, count in buildings.items():
        provision += data[building]["pop"] * count
    return provision
```

### Requirements

How many buildings are required for the desired population

```python
population = housing_desired["9x9"] * data["housing"]["9x9"]["pop"]
```

```python
farms_needed = population / data["food"]["wheat_farm"]["pop"]
print(f"{farms_needed=}")
```

```python
workshops_needed = total_houses / houses_per_workshop
print(f"{workshops_needed}")
```

### Provision

How many people the current number of buildings can sustain

```python
food_provided = provision(food_buildings, data["food"]) 
print(f"{food_provided=}")
```

```python
religion_provided = provision(religious_buildings, data["religion"])
print(f"{religion_provided=}")
```

```python
houses_provided = {
    "pottery": production_buildings["pottery"] * houses_per_workshop,
    "furniture": production_buildings["furniture"] * houses_per_workshop
}
print(f"{houses_provided=}")
```
